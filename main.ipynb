```python
import pandas as pd

data = pd.read_csv("netflix-dataset.csv")

print("\n===== Netflix Dataset =====\n")

print(data.head())
```

```python
print("\nDataset Shape:")
print(data.shape)
```

```python
print("\nColumns:\n")
print(data.columns)
```

```python
print("\n===== Content Type Count =====\n")

print(data["type"].value_counts())
```

```python
print("\n===== Top Genres =====\n")

print(data["listed_in"].value_counts().head(10))
```

```python
print("\n===== Top Release Years =====\n")

print(data["release_year"].value_counts().head(10))
```

```python
print("\n===== Top Countries =====\n")

print(data["country"].value_counts().head(10))
```

```python
movie_name = input("\nEnter movie title: ")

result = data[data["title"].str.contains(movie_name, case=False, na=False)]

print("\n===== Search Result =====\n")

print(result[["title", "type", "release_year"]])
```

```python
print("\n===== Missing Values =====\n")

print(data.isnull().sum())
```

```python
import matplotlib.pyplot as plt

type_count = data["type"].value_counts()

type_count.plot(kind="bar")

plt.title("Movies vs TV Shows")

plt.xlabel("Type")

plt.ylabel("Count")

plt.show()
```

```python
release_years = data["release_year"].value_counts().head(10)

release_years.plot(kind="bar")

plt.title("Top Release Years")

plt.xlabel("Year")

plt.ylabel("Number of Shows")

plt.show()
```
